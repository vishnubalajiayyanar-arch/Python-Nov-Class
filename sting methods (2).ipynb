```python
#string methods

#uppercase its captial our first value of string
a=("python")
print(a.upper())

#lowercase its lower case our first value of string
b=(a)
print(b.lower())

#capitalize its captialize our whole string values
print(a.capitalize())

#title its convert lower to upper and upper to lower our first value of string 
c=(a+b)
print(c.title())

#swapcase its convert whole string values in lower to upper and upper to lower
A=("python")
print((A.swapcase()),(c.swapcase()))

#strip its remove our spaces inside the string
r="     python program      "
t=(r.strip())
print("the spaces remove from",t," is our string")

#replace its relace particular sting is our values
print(r.replace("program","basic"))

#join its joint the all string 
g=["python","program","is","fun"]
print("".join(g))

#count its count a letter inside our integer values
f=[1,2,3,22,42,4,29,33]
count_2=f.count(2)
print(f"the letter of 2 is inside the f is{count_2} times in a.")

#find its find the letter inside our string values
j="hi progråm"
index=r.find(j)
print(f"the j is {j}start index:{index}")

#encode its encode the special charactervalue
x=(j.encode())
print(x)

#endswith its check end letter
y=j.endswith("m")
print(y,type(y))

#casefold its lowercase the whole string values
print(c.casefold())

#expandtaps its set tab size and remove tapsize in our string
L="p\ty\tt\th\tonprogram"
print(L)
J=(L.expandtabs(0))
print(J)

#format diffiret place holder values 
print("the expand tabs value of {}, is{}".format(L,J))

#format_map irs remove particular values from string in dictionary ina string values
Z={"name": " python","batch": " nov","year": " 2025"}
X="we scuccesfully contain{name} in this{batch} on {year}"
print(X.format_map(Z))

#index search the specified values in the sting and retrun the value
N=("name,value,data,value")
D=(N.index("name"))
print(D)

#isalnum,
K=("its very wrong bro")
I=(3,1,7,9.0,8.7)
print(K.isalnum())
print(K.isalpha())
```

```python
#isidentifier
x="demo"
y=x.isidentifier()
print(y)

#islower
z=x.islower()
print(z)

#isnumeric
t="547587"
n=t.isnumeric()
print(n)

#isprintabe
j="python november batch 2025 is @3 nice"
l=j.isprintable()
print(l)

#isspace
p=j.isspace()
print(p)

#isupper
q=j.isupper()
print(q)
```
