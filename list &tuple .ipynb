```python
#list

#append is add the values is list
a=[1,2,3,4]
a.append(5)
print(a)
```

```python
#clear
a=[1,2,3,4,5]
a.clear()
print(a)
```

```python
#copy 
a=[1,2,3,4,5]
b=a.copy()
print(b)
```

```python
#count
a=["python","weekend","batch"]
print(a.count("python"))

#extend
b=["2025","list","methods"]
a.extend(b)
print(a)

#index
c=a.index("methods")
print(c)

```

```python
#insert
a=["welcome","to","weekend","batch"]
a.insert(2,"python")
print(a)

#pop
a.pop(2)
print(a)

#remove
a.remove("to")
print(a)
```

```python
#reverse
a=[1,2,3,4,5]
a.reverse()
print(a)
```

```python
#sort
a=["welcome","to","python",]
a.sort()
print(a)
```

```python
#tuple
#count and Index its immutable 

#count
a=(1,2,3,4,5,5)
b=a.count(5)
print(b,type(a))
```

```python
#index
a=(1,2,3,4,5,7,8,9,0)
b=a.index(3)
print(b)
```
